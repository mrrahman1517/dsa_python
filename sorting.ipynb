```python
def quicksort(L, l, r):
    if r - l <= 1:
        return L
    (pivot, lower, upper) = (L[l], l + 1, l + 1)
    for i in range(l + 1, r):
        if L[i] > pivot:
            upper += 1
        else:
            (L[i], L[lower]) = (L[lower], L[i])
            lower += 1
            upper += 1
    (L[l], L[lower - 1]) = (L[lower - 1], L[l])
    lower -= 1
    quicksort(L, l, lower)
    quicksort(L, lower + 1, upper)
    return L
```

```python
# Define the qsort (quicksort) function recursively
def qsort(nums):
  # Base case: if the list has 0 or 1 element, it's already sorted
  if len(nums) <= 1:
    return nums

  # Choose the first element as the pivot
  pivot = nums[0]
  # Create a sublist of elements less than or equal to the pivot
  left = [num for num in nums[1:] if num <= pivot]
  # Create a sublist of elements greater than the pivot
  right = [num for num in nums[1:] if num > pivot]

  # Recursively sort the left and right sublists and combine them with the pivot
  return qsort(left) + [pivot] + qsort(right)
```

```python
# Define a function to test if an array is sorted in ascending order
def sort_tester(nums):
  # Iterate through the array up to the second-to-last element
  for i in range(len(nums)-1):
    # If any element is greater than the next element, the array is not sorted
    if nums[i] > nums[i+1]:
      return False
  # If the loop completes, the array is sorted
  return True
```
